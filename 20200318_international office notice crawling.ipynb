게시글 URL, 날짜, 제목 크롤링

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.skku.edu/skku/international/globalComm/international01.do?'
```

```python
office_notice = []

for idx in range(0, 41, 10):
    params = {
        'mode' : 'list',
        'articleLimit' : '10',
        'article.offset' : idx
    }
    
    resp = requests.get(url, params)
    soup = BeautifulSoup(resp.content, 'lxml')
    
    # no
    no_td_tag = soup.select('div.board-wrap.board-qa.table_scrollWrap > table > tbody > tr > td:nth-child(1)')
    
    # title, url
    a_tag = soup.select('td.left > a')
    
    # date
    date_td_tag = soup.select('div.board-wrap.board-qa.table_scrollWrap > table > tbody > tr > td:nth-child(3)')
    
    for i in range(10):
        result_dic = {
            'no' : no_td_tag[i].text.strip(),
            'title' : a_tag[i].text.strip(),
            'date' : date_td_tag[i].text,
            'url' : 'https://www.skku.edu/skku/international/globalComm/international01.do' + a_tag[i]['href']
        }
        
        office_notice.append(result_dic)
```

```python
office_notice
```

```python
import csv

csv_columns = ['no', 'title', 'date', 'url']

try:
    with open("office_notice_data.csv", 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for data in office_notice:
            writer.writerow(data)
except IOError:
    print("I/O error")
```
