### 대학별 이미지주소 크롤링

```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
univ_list = []

with open('./univ_data.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for data in reader:
        if (data['name'] != ''):
            univ_list.append(data['name'])
        
univ_list
```

```python
url = 'https://www.google.com/search?'
```

```python
univ_image_url = {}

for idx, uni in enumerate(univ_list):
    print('크롤링 중 ... {}'.format(idx))
    params = {
        'tbm' : 'isch',
        'q' : uni
    }
    
    resp = requests.get(url, params)
    soup = BeautifulSoup(resp.content, 'lxml')
    
    img_tag = soup.find('img', {"style" : "border:1px solid #ccc;padding:1px"})
    img_src = img_tag['src']

    univ_image_url[uni] = img_src
```

```python
f = open("univ_img_src.txt", 'wb')
pickle.dump(univ_image_url, f) # pickling
f.close()
```
