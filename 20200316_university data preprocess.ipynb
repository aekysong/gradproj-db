```python
import csv
from collections import defaultdict
import pickle
```

### NLP 결과로 나온 대학별 대표 키워드 추가하기

```python
with open('result_keyword.txt', 'rb') as file:
    nlp_keyword = pickle.load(file)
    
nlp_keyword[''] = ''
nlp_keyword
```

```python
for key in nlp_keyword.keys():
    temp = []
    for keyword, tfidf in nlp_keyword[key]:
        temp.append(keyword)
    nlp_keyword[key] = temp
nlp_keyword
```

### 구글 이미지 검색에서 크롤링해서 가져온 대학별 이미지 url

```python
with open('univ_img_src.txt', 'rb') as file:
    univ_image_url = pickle.load(file)
        
univ_image_url[''] = ''
univ_image_url
```

### 대학별 만족도 평균(univ_satisfaction) 및 몇 명이 해당 대학에 파견됐었는지(univ_count) 계산

```python
univ_satisfaction = defaultdict(lambda: 0)
univ_count = defaultdict(lambda: 0)
univ_major = defaultdict(lambda: [])

with open('./report_data.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for data in reader:
        univ_satisfaction[data['university']] += int(data['final_satisfaction'])
        univ_count[data['university']] += 1
        if data['major'] not in univ_major[data['university']] and data['major'] != '':
            univ_major[data['university']].append(data['major'])

for key in univ_satisfaction:
    univ_satisfaction[key] = round(univ_satisfaction[key] / univ_count[key], 2)
    
univ_satisfaction
```

### 국제처에서 올려준 파견대학 리스트 엑셀 파일에 있던 정보 가져오기

```python
univ_data = []

for i in range(len(pre_univ_list)):
    try:
        result_dict = {
            'name': pre_univ_list[i]['\ufeff대학명'],
            'nation': pre_univ_list[i]['국가'],
            'remarks': pre_univ_list[i]['비고사항'],
            'website': pre_univ_list[i]['홈페이지 제공 여부'],
            'available_number': pre_univ_list[i]['선발인원(교환학생)'],
            'language_condition': pre_univ_list[i]['어학자격기준'],
            'one_year_available': pre_univ_list[i]['1년 파견 가능 여부'],
            'course_restriction': pre_univ_list[i]['수강 제한 영역 여부                                                                                                   (Are there any specific areas of study in which exchange students cannot take courses?)'],
            'not_major_course': pre_univ_list[i]['전공 이외의 수업 수강 가능 여부                                                                 (Do you allow students to take courses outside of their applied major?)'],
            'keyword': nlp_keyword[pre_univ_list[i]['\ufeff대학명']],
            'satisfaction' : univ_satisfaction[pre_univ_list[i]['\ufeff대학명']],
            'went_number': univ_count[pre_univ_list[i]['\ufeff대학명']],
            'went_major': univ_major[pre_univ_list[i]['\ufeff대학명']],
            'image_url' : univ_image_url[pre_univ_list[i]['\ufeff대학명']]
        }
    except:
        result_dict = {
            'name': pre_univ_list[i]['\ufeff대학명'],
            'nation': pre_univ_list[i]['국가'],
            'remarks': pre_univ_list[i]['비고사항'],
            'website': pre_univ_list[i]['홈페이지 제공 여부'],
            'available_number': pre_univ_list[i]['선발인원(교환학생)'],
            'language_condition': pre_univ_list[i]['어학자격기준'],
            'one_year_available': pre_univ_list[i]['1년 파견 가능 여부'],
            'course_restriction': pre_univ_list[i]['수강 제한 영역 여부                                                                                                   (Are there any specific areas of study in which exchange students cannot take courses?)'],
            'not_major_course': pre_univ_list[i]['전공 이외의 수업 수강 가능 여부                                                                 (Do you allow students to take courses outside of their applied major?)'],
            'keyword': None,
            'satisfaction' : univ_satisfaction[pre_univ_list[i]['\ufeff대학명']],
            'went_number': univ_count[pre_univ_list[i]['\ufeff대학명']],
            'went_major': univ_major[pre_univ_list[i]['\ufeff대학명']],
            'image_url' : univ_image_url[pre_univ_list[i]['\ufeff대학명']]
        }
    
    univ_data.append(result_dict)
```

### CSV 파일로 내보내기

```python
csv_columns = ['name', 'nation', 'remarks', 'website', 'available_number', 
              'language_condition', 'one_year_available', 'course_restriction',
               'not_major_course', 'keyword', 'satisfaction', 'went_number',
              'went_major', 'image_url']

try:
    with open("univ_data.csv", 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for data in univ_data:
            writer.writerow(data)
except IOError:
    print("I/O error")
```
