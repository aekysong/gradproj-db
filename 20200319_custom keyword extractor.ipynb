## 1. TF-IDF (문서별)

```python
from konlpy.tag import Okt

def f(t, d):
    return d.count(t)

def tf(t, d):
    return 0.5 + 0.5*f(t,d)/max([f(w,d) for w in d])

def idf(t, D):
    numerator = len(D)
    denominator = 1 + len([ True for d in D if t in d])
    return log10(numerator/denominator)

def tfidf(t, d, D):
    return tf(t,d)*idf(t, D)

def okt_tokenize(d):
    okt=Okt()
    tokens = okt.pos(d, stem=True) # 단어 기준
    result_tokens = []
    for token in tokens: # 한 글자인 단어 제외, 명사, 동사, 형용사만 키워드로 포함
        if ((len(token[0]) > 1) and (token[1] == 'Noun' or token[1] == 'Verb' or token[1] == 'Adjective')):
            result_tokens.append(token[0])
    return result_tokens

def tfidfScorer(D):
    tokenized_D = [okt_tokenize(d) for d in D] # 토큰화 후
    result = []
    for d in tokenized_D:
        result.append([(t, tfidf(t, d, tokenized_D)) for t in d]) # tf-idf 값 계산
    return result
```

```python
univ_tfidf_1 = {}

for idx, key in enumerate(univ_final_comment.keys()):
    print('키워드 추출 중 ... {}'.format(idx + 1))
    corpus = univ_final_comment[key] # 한 대학의 레포트들 리스트
    temp = []
    for s in tfidfScorer(corpus):
        temp += s[:10]
    univ_tfidf_1[key] = sorted(set(temp), key=lambda x:x[1], reverse=True)[:10] # 레포트가 아니라 대학 기준으로 키워드 모으기
```

## 2. TF-IDF (대학별 TF / 문서별 IDF)

```python
def f(t, d): # 레포트가 아니라 한 대학 전체 레포트에서 t가 들어있는 개수 구하기
    return d.count(t)

def tf(t, d): # TF 산출 방식 중 증가 빈도 사용
    return 0.5 + 0.5*f(t,d)/max([f(w,d) for w in d])

def idf(t, D):
    numerator = len(D)
    denominator = 1 + len([ True for d in D if t in d])
    return log10(numerator/denominator)

def tfidf(t, join_D, D): # join_D는 한 대학 전체 레포트
    return tf(t, join_D)*idf(t, D)

def tfidfScorer(D):
    tokenized_D = [okt_tokenize(d) for d in D] # [[ , , , ...], [], [], ...]
    unique_t = []
    for d in tokenized_D:
        for t in d:
            if t not in unique_t:
                unique_t.append(t)
    join_D = ' '.join(D)
    join_D = okt_tokenize(join_D)
    result = [(t, tfidf(t, join_D, tokenized_D)) for t in unique_t]
    return result
```

```python
univ_tfidf_2 = {}

for idx, key in enumerate(univ_final_comment.keys()):
    print('키워드 추출 중 ... {}'.format(idx + 1))
    corpus = univ_final_comment[key]
    univ_tfidf_2[key] = sorted(set(tfidfScorer(corpus)), key=lambda x:x[1], reverse=True)[:10]
```

## 3. TF-IDF (대학별)

```python
def f(t, d): 
    return d.count(t)

def tf(t, d): # TF 산출 방식 중 증가 빈도 사용
    return 0.5 + 0.5*f(t,d)/max([f(w,d) for w in d])

def idf(t, D):
    numerator = len(D)
    denominator = 1 + len([ True for d in D if t in d])
    return log10(numerator/denominator)

def tfidf(t, d, D): 
    return tf(t, d)*idf(t, D)

def tfidfScorer(D, G):
    tokenized_D = okt_tokenize(D)
    result = []
    for t in tokenized_D:
        result.append((t, tfidf(t, tokenized_D, G)))
    return result
```

```python
univ_tfidf_3 = {}

total_univ = list(univ_final_comment.values())
tokenized_G = []
for value in total_univ:
    value = ' '.join(value)
    tokenized_G.append(okt_tokenize(value))

for idx, key in enumerate(univ_final_comment.keys()):
    print('키워드 추출 중 ... {}'.format(idx + 1))
    corpus = univ_final_comment[key]
    corpus = ' '.join(corpus)
    univ_tfidf_3[key] = sorted(set(tfidfScorer(corpus, tokenized_G)), key=lambda x:x[1], reverse=True)[:10]
```

## 5. TextRank
https://github.com/lovit/textrank   
오픈소스 활용

```python
from textrank import KeywordSummarizer
```

```python
from konlpy.tag import Komoran

komoran = Komoran()

def komoran_tokenize(sent):
    words = komoran.pos(sent, join=False)
    result = []
    for w in words:
        if ((len(w[0]) > 1) and (w[1] == 'NN' or w[1] == 'VA' or w[1] == 'VV')):
            result.append(w[0])
    return result
```

```python
univ_textrank = {}

keyword_extractor = KeywordSummarizer(
    tokenize = okt_tokenize,
    window = -1,
    verbose = False
)
    
for idx, key in enumerate(univ_final_comment.keys()):
    print('키워드 추출 중 ... {}'.format(idx + 1))
    corpus = univ_final_comment[key]
    try:
        keywords = keyword_extractor.summarize(corpus, topk=30)
    except:
        keywords = []
    univ_textrank[key] = keywords[:10]
```
