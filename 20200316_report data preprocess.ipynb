```python
for i in range(len(report_data)):
    univ_flag = False
    major_flag = False
    
    # university name
    for k in range(len(univ_list)):
        if report_data[i]['university'].lower() in univ_list[k].lower():
            univ_flag = True
            report_data[i]['pre_university'] = report_data[i]['university']
            report_data[i]['university'] = univ_list[k]
            break
        # 예외 처리
        if 'vrije' in report_data[i]['university'].lower(): 
            univ_flag = True
            report_data[i]['pre_university'] = report_data[i]['university']
            report_data[i]['university'] = 'Vrije Universiteit Amsterdam'
            break
            
    if not univ_flag:
        report_data[i]['university'] = '' # bulk_insert에서 거르기
        
    # major_name
    report_data[i]['major'] = report_data[i]['major'].split()[0].split('(')[0].split('/')[0].split(',')[0]
    
    for k in range(len(major_list)):
        if report_data[i]['major'] in major_list[k]:
            major_flag = True
            report_data[i]['major'] = major_list[k]
            print(report_data[i]['major'])
            break
            
    if not major_flag:
        report_data[i]['major'] = '' # bulk_insert에서 거르기
    
    # dorm_satisfaction
    if (('HYPERLINK' in report_data[i]['dorm_satisfaction']) or 
        ('{기숙사 및 기타 숙소 관련 정보를 얻을 수 있는 웹사이트)' in report_data[i]['dorm_satisfaction']) or
        ('www' in report_data[i]['dorm_satisfaction'])):
        report_data[i]['dorm_satisfaction'] = ''
```

```python
csv_columns = ['pre_university', 'university', 'student_id', 'department', 'major', 
               'ex_period', 'visa_type', 'visa_period', 'visa_process',
              'dorm_name', 'dorm_location', 'dorm_cost', 'dorm_satisfaction',
              'dorm_comment', 'lecture_comment', 'lecture_grading', 'culture_activity',
              'final_satisfaction', 'final_comment']

try:
    with open("report_data.csv", 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for data in report_data:
            writer.writerow(data)     
except IOError:
    print("I/O error")
```
